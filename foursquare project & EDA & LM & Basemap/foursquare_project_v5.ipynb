### Data Frame Definition for Venue Categories only for Food Places

```python
import pandas as pd
import numpy as np
dfCategory=pd.DataFrame(columns=["Name","PluralName","ShortName"])
foodKey="4d4b7105d754a06374d81259"
def add_row(df, subcategory):
    name = subcategory["name"]
    pluralname = subcategory["pluralName"]
    shortname = subcategory["shortName"]
    df.loc[subcategory["id"]] = [name,pluralname,shortname]
    
for row in venueCategories["response"]["categories"]:
    if row["id"]==foodKey:
        for subrow in row["categories"]:
                add_row(dfCategory, subrow)

dfCategory.to_csv("Categories.csv",index_label="id",encoding="utf-8")
dfCategory
```

### Venues Dataframe Definition and Data Conversion and Initial Cleansing

```python
import pandas as pd
import numpy as np

dfVenue = pd.DataFrame(columns=["Name","Lat","Lon","Distance","CC","City","State","Country","CatId","CatName","CatPluralName","CatShortName","PostalCode","isVerified","StatsCheckinsCount","StatsUsersCount","StatsTipCount",
                                "PriceTier","PriceMessage","LikesCount","Rating","RatingSignals","RatingColor","hasMenu","MenuUrl","MenuMobileUrl","SpecialsCount","PhotosCount"])

def add_row(df, venue):
    name=venue["name"]
    if venue["location"].get("postalCode"):
        postalCode=venue["location"]["postalCode"]
    else:
        postalCode=0
    lat=venue["location"]["lat"]
    lon=venue["location"]["lng"]
    distance=venue["location"]["distance"]
    if venue["location"].get("cc"):
        cc=venue["location"]["cc"]
    else:
        cc=""
    if venue["location"].get("city"):
        city=venue["location"]["city"]
    else:
        city=""
    if venue["location"].get("state"):
        state=venue["location"]["state"]
    else:
        state=""
    if venue["location"].get("country"):
        country=venue["location"]["country"]
    else:
        country=""
    
    verified=venue["verified"]
    statsCheckinsCount=venue["stats"]["checkinsCount"]
    statsUsersCount=venue["stats"]["usersCount"]
    statsTipCount=venue["stats"]["tipCount"]
    if venue.get("price"):
        priceTier=venue["price"]["tier"]
        priceMessage=venue["price"]["message"]
    else:
        priceTier=0
        priceMessage=''
    if venue.get("hasMenu"):
        hasMenu=venue["hasMenu"]
    else:
        hasMenu=False
    
    if venue.get("likes"):
        likesCount=venue["likes"]["count"]
    else:
        likesCount=0
    if venue.get("rating"):
        rating=venue["rating"]
    else:
        rating=0
    if venue.get("ratingColor"):
        ratingColor=venue["ratingColor"]
    else:
        ratingColor=''
    if venue.get("ratingSignals"):
        ratingSignals=venue["ratingSignals"]
    else:
        ratingSignals=0
    if venue.get("menu"):
        menuUrl=venue["menu"]["url"]
        menuMobileUrl=venue["menu"]["mobileUrl"]
    else:
        menuUrl=''
        menuMobileUrl=''
    
    specialsCount=venue["specials"]["count"]
    photosCount=venue["photos"]["count"]

    for cat in venue["categories"]:
        catname = cat["name"]
        pluralname = cat["pluralName"]
        shortname = cat["shortName"]
        
        isprimary=cat["primary"]
        
        if isprimary:

            df.loc[venue["id"]] = [name,lat,lon,distance,cc,city,state,country,cat["id"],catname,pluralname,shortname,postalCode,verified,statsCheckinsCount,statsUsersCount,statsTipCount,priceTier,priceMessage,
                                   likesCount,rating,ratingSignals,ratingColor,hasMenu,menuUrl,menuMobileUrl,specialsCount,photosCount]
            break;
        elif cat["id"] in dfCategory.index:
            df.loc[venue["id"]] = [name,lat,lon,distance,cc,city,state,country,cat["id"],catname,pluralname,shortname,postalCode,verified,statsCheckinsCount,statsUsersCount,statsTipCount,priceTier,priceMessage,
                                   likesCount,rating,ratingSignals,ratingColor,hasMenu,menuUrl,menuMobileUrl,specialsCount,photosCount]
            break;
        
```

### Deriving Some Discretized Variables that can be used as an independent variable in linear regression

```python

dfRestaurant["PhotosCountCat"]=pd.Categorical(dfRestaurant.PhotosCount)
dfRestaurant["DistanceCat"]=pd.Categorical(dfRestaurant.Distance)
def makeInt(x):
    return int(x)

dfRestaurant["RatingSignalsCat"]=pd.Categorical(dfRestaurant.RatingSignals/100)
dfRestaurant["RatingCat"]=pd.Categorical(dfRestaurant.Rating.map(makeInt))
```

### Bivariant Linear Regressions

```python
import statsmodels.formula.api as sm
from IPython.core.display import HTML
 

def short_summary(est):
    return HTML(str(est.summary().tables[0].as_html()))


# Fit and summarize OLS model
mod = sm.ols(formula="Rating ~ PhotosCountCat", data=dfRestaurant)
res = mod.fit()
ols_ci = res.conf_int().ix['PhotosCountCat'].tolist()
ols = dict(a = res.params['Intercept'],
           b = res.params['PhotosCountCat'])
plt.figure(1,figsize=(10,6)) 

x = np.linspace(dfRestaurant["PhotosCountCat"].min(), dfRestaurant["PhotosCountCat"].max(), 100)
get_y = lambda a, b: a + b * x 
y = get_y(ols['a'], ols['b'])


plt.scatter(dfRestaurant.PhotosCount, dfRestaurant.Rating,  alpha=.6, color='b', s=50)
plt.xlabel("Number of Photos")
plt.ylabel("Food Places Rating")
plt.title("Linear Regression Model Rating ~ PhotosCountCat")
plt.plot(x, y, color='red', label='OLS')
short_summary(res)
```

Number of Photos does not have an important impact on customer's rating behaviour. 

```python
plt.clf()
mod = sm.ols(formula="Rating ~ DistanceCat", data=dfRestaurant)
res = mod.fit()
ols_ci = res.conf_int().ix['DistanceCat'].tolist()
ols = dict(a = res.params['Intercept'],
           b = res.params['DistanceCat'])

rating_linspace = np.linspace(dfRestaurant["DistanceCat"].min(), dfRestaurant["DistanceCat"].max(), 100)

get_y = lambda a, b: a + b * x 
y = get_y(ols['a'], ols['b'])
plt.figure(1,figsize=(10,6)) 
plt.subplot(1,1,1)
plt.scatter(dfRestaurant.Distance,dfRestaurant.Rating, lw=0, alpha=.6, color='b', s=50)
plt.xlabel("Distance of a Food Place from the City Center -meters-")
plt.ylabel("Rating")
plt.title("Linear Regression Model Rating ~ Distance Categories")
plt.plot(x, y, color='red', label='OLS')
short_summary(res)
```

```python
import math
class BoundingBox(object):
    def __init__(self, *args, **kwargs):
        self.lat_min = None
        self.lon_min = None
        self.lat_max = None
        self.lon_max = None


def get_bounding_box(latitude_in_degrees, longitude_in_degrees, half_side_in_km):
    assert half_side_in_km > 0
    assert latitude_in_degrees >= -180.0 and latitude_in_degrees  <= 180.0
    assert longitude_in_degrees >= -180.0 and longitude_in_degrees <= 180.0

#half_side_in_km = half_side_in_miles * 1.609344
    lat = math.radians(latitude_in_degrees)
    lon = math.radians(longitude_in_degrees)

    radius  = 6370.997
    # Radius of the parallel at given latitude
    parallel_radius = radius*math.cos(lat)

    lat_min = lat - half_side_in_km/radius
    lat_max = lat + half_side_in_km/radius
    lon_min = lon - half_side_in_km/parallel_radius
    lon_max = lon + half_side_in_km/parallel_radius
    rad2deg = math.degrees

    box = BoundingBox()
    box.lat_min = rad2deg(lat_min)
    box.lon_min = rad2deg(lon_min)
    box.lat_max = rad2deg(lat_max)
    box.lon_max = rad2deg(lon_max)

    return (box)
box=get_bounding_box(maplat, maplon, float(radius)/1000)

#zooming in with a ratio of aspect because when directly variabled maplat and maplon are provided maximum zoom level is not 
#sufficient to have a comprensible map.
dlatdeg=box.lat_max-box.lat_min
dlondeg=box.lon_max-box.lon_min
zoomy=11
zoomx=11
#The y_adj adjustment below is just to move the bounding bow a bşt to north 
#because central latitude & longitude of manhattan in Foursquare is just arranged to show Financial Center as the center of NYC. 
#In order to show all island including Harlem, Yoh Yoh!
y_adj=0.019
urcrnrlat=(box.lat_min+box.lat_max)/2+dlatdeg/(2*zoomy)+y_adj
llcrnrlat=(box.lat_min+box.lat_max)/2-dlatdeg/(2*zoomy)+y_adj
urcrnrlon=(box.lon_min+box.lon_max)/2+dlondeg/(2*zoomx)
llcrnrlon=(box.lon_min+box.lon_max)/2-dlondeg/(2*zoomx)
```

Bounding Box is used to have a robust frame that protects the scale of the map
